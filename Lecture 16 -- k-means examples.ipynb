```python
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, cluster_labels = make_blobs(n_samples=500, centers=3, cluster_std=0.3, random_state=0)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster_labels)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('2-dimensional Gaussians')
```

```python
cluster_labels
```

```python
kmeans = KMeans(n_clusters=3, init='random', max_iter=300, random_state=0)
pred_cluster_labels = kmeans.fit_predict(X)
```

```python
pred_cluster_labels
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster_labels)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', s=50)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('2-dimensional Gaussians')
```
